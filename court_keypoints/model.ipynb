```python

import os

labels_dir = "reloc2-1/train/labels"
images_dir = "reloc2-1/train/images"

# Check for missing or empty label files
missing_labels = []
empty_labels = []
for img_file in os.listdir(images_dir):
	if img_file.endswith(('.jpg', '.png', '.jpeg')):
		label_file = os.path.splitext(img_file)[0] + ".txt"
		label_path = os.path.join(labels_dir, label_file)
		if not os.path.exists(label_path):
			missing_labels.append(label_file)
		elif os.path.getsize(label_path) == 0:
			empty_labels.append(label_file)

print("Missing label files:", missing_labels)
print("Empty label files:", empty_labels)
```

```python
import os

# Paths to label directories
train_labels_dir = "/home/marcos/Vision/project-2/court_keypoints/basketball-1/train/labels"
valid_labels_dir = "/home/marcos/Vision/project-2/court_keypoints/basketball-1/valid/labels"
test_labels_dir = "/home/marcos/Vision/project-2/court_keypoints/basketball-1/test/labels"

# Function to check label files
def check_labels(labels_dir):
    missing_labels = []
    empty_labels = []
    invalid_labels = []

    for label_file in os.listdir(labels_dir):
        label_path = os.path.join(labels_dir, label_file)

        # Check if the label file is empty
        if os.path.getsize(label_path) == 0:
            empty_labels.append(label_file)
            continue

        # Check the format of the label file
        with open(label_path, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    invalid_labels.append(label_file)
                    break
                try:
                    class_id = int(parts[0])
                    coords = list(map(float, parts[1:]))
                    if not all(0 <= coord <= 1 for coord in coords):
                        invalid_labels.append(label_file)
                        break
                except ValueError:
                    invalid_labels.append(label_file)
                    break

    return missing_labels, empty_labels, invalid_labels

# Check all label directories
for labels_dir, name in zip([train_labels_dir, valid_labels_dir, test_labels_dir], ["train", "valid", "test"]):
    missing, empty, invalid = check_labels(labels_dir)
    print(f"\n{name.capitalize()} Labels Check:")
    print(f"Missing labels: {missing}")
    print(f"Empty labels: {empty}")
    print(f"Invalid labels: {invalid}")
```
